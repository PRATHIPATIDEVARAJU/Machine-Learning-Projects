```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data['Month'] = le.fit_transform(data['Month'])
data['Region'] = le.fit_transform(data['Region'])
data['Main_Category'] = le.fit_transform(data['Main_Category'])
data['Sub_Category'] = le.fit_transform(data['Sub_Category'])
data['OS'] = le.fit_transform(data['OS'])
data['Primary Use'] = le.fit_transform(data['Primary Use'])
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(['OS'], axis=1)
y = data['OS']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

print("x_train:", x_train.shape)
print("x_test:", x_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.svm import SVC

model = SVC(kernel='linear',random_state=1)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.4f}")
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)

print(f" Accuracy of R2_Score = {r2}")
```
